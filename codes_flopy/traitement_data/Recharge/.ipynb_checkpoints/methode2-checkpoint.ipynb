```python
import geopandas as gp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import rasterio 
import flopy as fp
from flopy.utils import Raster
```

```python
def resample_rast(rast_path,grid,band=1):
    arr = Raster.load(rast_path)
    arr = arr.resample_to_grid(grid.xcellcenters,grid.ycellcenters,band)
    return arr
```

```python
# load des diff√©rents rasters
z_rast = "../../data/recharge/zones_methode2/zones_rast.tif"
rast_rch = "../../data/recharge/recharge_p.tif"

# extract boundaries data from surfaces
MNT_path = "../../data/surfaces_geoter/TOPO  (0 en mer).ASC"
x0,y0,x1,y1 = rasterio.open(MNT_path).bounds
x1 -= 3e4
# set resolution of the array
res = 50
Lx = x1-x0
Ly = y1-y0
nrow = int(Ly/res)
ncol = int(Lx/res)

#create grid with flopy
grid = fp.discretization.StructuredGrid(delc=res*np.ones([nrow]),delr=res*np.ones([ncol]),xoff=x0,yoff=y0)

#resample zones with the grid
zones = resample_rast(z_rast,grid)
```

```python
# liste of irrigation coefficients by zones
coeff = [2,2,2,0,2,2,2,2]
plt.imshow(zones)
plt.colorbar()
```

```python
# create a new array with coeff apllied
mult = zones.copy()
for i in range(len(coeff)):
    mult[zones==i+1]=coeff[i]
plt.imshow(mult)
```

```python
rch = resample_rast(rast_rch,grid)
rch[rch<0]=np.mean(rch[rch>0])
plt.imshow(rch)
plt.colorbar()
```

```python
#calculate the recharge from irrigation
fig = plt.figure(figsize=(12,12))
irr_rch = mult*rch
plt.imshow(irr_rch)
plt.colorbar()
plt.title("recharge par irrigation")
```

```python
irr_rch[irr_rch<0] = 0
```

```python
def arr2ascii(arr,filename,x0,y0,res,nodata=-9999):
    ncol = arr.shape[1]
    nrow = arr.shape[0]
    with open(filename,"w") as file:
        file.write("ncols {}\n".format(ncol))
        file.write("nrows {}\n".format(nrow))
        file.write("xllcorner {}\n".format(x0))
        file.write("yllcorner {}\n".format(y0))
        file.write("cellsize {}\n".format(res))
        file.write("nodata_value {}\n".format(nodata))
        for irow in range(nrow):
            for icol in range(ncol):
                file.write(str(arr[irow,icol])+" ")
```

```python
arr2ascii(irr_rch,"../../data/recharge/rast_irr.asc",x0,y0,res)
```
