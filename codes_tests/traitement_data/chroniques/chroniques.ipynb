```python
def coor_convert(x,y,epsgin,epsgout):
    from pyproj import Proj, transform
    inproj = Proj(init="epsg:{}".format(epsgin))
    outproj = Proj(init="epsg:{}".format(epsgout))
    xp,yp = transform(inproj,outproj,x,y)
    return xp,yp
```

```python
a["X_L93"]=coor_convert(a.X_WGS84.values,a.Y_WGS84.values,"4326","2154")[0]
a["Y_L93"]=coor_convert(a.X_WGS84.values,a.Y_WGS84.values,"4326","2154")[1]
```
