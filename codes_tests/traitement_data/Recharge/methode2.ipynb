```python
import geopandas as gp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import rasterio 
import flopy as fp
from flopy.utils import Raster
```

```python
def resample_rast(rast_path,band=1):
    arr = Raster.load(rast_path)
    arr = arr.resample_to_grid(grid.xcellcenters,grid.ycellcenters,band)
    return arr
```

```python
# load des diff√©rents rasters

z_rast = "../../data/recharge/zones_methode2/zones_rast.tif"
rast_rch = "../../data/recharge/recharge_p.tif"
grid = fp.discretization.StructuredGrid(delc=300*np.ones([170]),delr=300*np.ones([173]),xoff=664480,yoff=6153022)
zones = resample_rast(z_rast)
```

```python
# liste des coefficients multiplicateurs en fonction de la zone (un num√©ro diff√©rent pour chaque)
coeff = [4,3,3,3,4,3,3,3]
plt.imshow(zones)
plt.colorbar()
```

```python
mult = zones.copy()
for i in range(len(coeff)):
    mult[zones==i+1]=coeff[i]
plt.imshow(mult)
```

```python
rch = resample_rast(rast_rch)
plt.imshow(rch,vmin=0)
plt.colorbar()
```

```python
#calculate the recharge from irrigation
fig = plt.figure(figsize=(12,12))
irr_rch = mult*rch
plt.imshow(irr_rch)
plt.colorbar()
plt.title("recharge par irrigation")
```
