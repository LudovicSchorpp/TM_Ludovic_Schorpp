This notwbook exploit the ability to resample mnt file to a specific grid, can be really slow if the mnt is extremely precise

```python
model_dir = "working"
model_name="mnt_test"
exe_name="../../exe.mf6"
R_path="../data/limiteModeleRoussillon_poly.shp" # path to the shp of the aquifer
MNT_path= "../../../../../MNT_50.tif"
MNT_path_asci = "../../data/txt_couches/s1_mnt.txt"
```

```python
x0 = 660000
y0 = 6150000
x1 = 706000
y1 = 6205000

Lx = x1-x0
Ly = y1-y0
nlay=1
ncol = 100
nrow = 100
delr = np.ones(ncol)*(Lx/ncol)
delc = np.ones(nrow)*(Ly/nrow)
botm = -100
top = 100
idomain = np.zeros((nrow*ncol))
```

```python
MNT = Raster.load(MNT_path)
MNT_asci =  Raster.load(MNT_path_asci)
```

```python
MNT.bounds
```

```python
grid = fp.discretization.StructuredGrid(delc,delr,xoff=x0,yoff=y0)
```

```python
fig, ax = plt.subplots(1,1,figsize=(15,15))
ax.set_aspect("equal")

MNT_asci.plot(ax=ax)
grid.plot(ax=ax,alpha=0.5)
```

```python
dem_data = MNT.resample_to_grid(grid.xcellcenters,
                                grid.ycellcenters,
                                band = MNT.bands[0],
                                method="nearest"
                                )

fig = plt.subplots(figsize=(10,10))
plt.imshow(dem_data)
plt.colorbar()
```

```python
dem_data_asci = MNT_asci.resample_to_grid(grid.xcellcenters,
                                grid.ycellcenters,
                                band = MNT.bands[0],
                                method="nearest"
                                )
```

```python
dem_data_asci[dem_data_asci==-9999]=0
```

```python
fig = plt.figure(figsize=(12,12))
g=plt.imshow(dem_data - dem_data_asci)
plt.colorbar(g)
g.set_clim(-10,30)
```
